```python
import numpy as np
import pandas as pd

import scanpy as sc

import matplotlib.pyplot as plt
import seaborn as sns

import myutils

import os

sc.settings.verbosity = 3
myutils.set_figure_params()
```

```python
data_meta = pd.read_excel("datasheet.xlsx")
data_meta.head(3)
```

```python
adatas = []
for i in range(data_meta.shape[0]):
    adata = sc.read_mtx(data_meta.loc[i,"path"]+'/matrix.mtx.gz')
    adata= adata.T
    adata.obs_names = pd.read_csv(data_meta.loc[i,"path"]+"/barcodes.tsv.gz",sep="\t",header=None)[0].tolist()
    adata.var_names = pd.read_csv(data_meta.loc[i,"path"]+"/features.tsv.gz",sep="\t",header=None)[0].tolist()
    adata.obs['cell_id'] = [data_meta.loc[i,"sample"]+"_"+barcode for barcode in adata.obs_names]
    adata.obs_names = adata.obs['cell_id']

    for col in data_meta.columns:
        adata.obs[col] = data_meta.loc[i,col]

    adata.var_names_make_unique()
    print(adata)
    adatas.append(adata)
    
adata = sc.concat(adatas)
print(adata)
adata.write_h5ad("1.scRNA_merge_raw.h5ad")
del adatas, data_meta
```
