```python
select_pairs = pd.read_excel("5.3.Interaction_pairs_0707.xlsx")
select_pairs['interaction_name'] = ['AREG_EGFR', 'HBEGF_EGFR', 'IGFBP3_TMEM219', 'JAG1_NOTCH2', 
                                    'VEGFA_NRP1', 'PLAU_PLAUR', 'VEGFB_NRP1', 'TGFB1_TGFBR3', 'GAS6_AXL', 'TGFA_EGFR', 'JAG2_NOTCH2', 'EFNB1_EPHB3', 
                                    'EFNB1_EPHB6', 'EDN1_EDNRA', 'SEMA3C_NRP2', 'TENM2_ADGRL2', 'EDN1_EDNRB', 'NRG1_ERBB3', 'PTPRD_IL1RAP']
```

```python
cpdb_results = {}
cpdb_results['deconvoluted'] = pd.read_csv("5.1.CellPhoneDB/results/method2_withScore/statistical_analysis_deconvoluted_07_21_2025_161848.txt",sep="\t")
cpdb_results['deconvoluted_percents'] = pd.read_csv("5.1.CellphoneDB/results/method2_withScore/statistical_analysis_deconvoluted_percents_07_21_2025_161848.txt",sep="\t")
cpdb_results['pvalues'] = pd.read_csv("5.1.CellphoneDB/results/method2_withScore/statistical_analysis_pvalues_07_21_2025_161848.txt",sep="\t")
cpdb_results['means'] = pd.read_csv("5.1.CellphoneDB/results/method2_withScore/statistical_analysis_means_07_21_2025_161848.txt",sep="\t")
cpdb_results['significant_means'] = pd.read_csv("5.1.CellphoneDB/results/method2_withScore/statistical_analysis_significant_means_07_21_2025_161848.txt",sep="\t")
cpdb_results['interaction_scores'] = pd.read_csv("5.1.CellphoneDB/results/method2_withScore/statistical_analysis_interaction_scores_07_21_2025_161848.txt",sep="\t")
cpdb_results.keys()
```

```python
tmp = ["Spi-1","Spi-2","Spi-EMT","Bas","Bas-pro","Bas-EMT","Granu"]
selective_senders_scar = [i+"_Scar/keloid" for i in tmp]
selective_senders_NS = [i+"_NS" for i in tmp]
tmp = ["Myofibroblast","Fibroblast (non-myo)"]
selective_receivers_scar = [i+"_Scar/keloid" for i in tmp]
selective_receivers_NS = [i+"_NS" for i in tmp]
use_pairs_scar = []
use_pairs_NS = []
use_pairs = []
for i in selective_senders_scar:
    for j in selective_receivers_scar:
        use_pairs_scar.append(i+"|"+j)
for i in selective_senders_NS:
    for j in selective_receivers_NS:
        use_pairs_NS.append(i+"|"+j)
for i in range(len(use_pairs_scar)):
    use_pairs.append(use_pairs_scar[i])
    use_pairs.append(use_pairs_NS[i])
print(use_pairs)
```

```python
import numpy as np
```

```python
cpdb_results['means'].index = cpdb_results['means']['interacting_pair']
use_means = cpdb_results['means'].loc[select_pairs['interaction_name']]
use_means = use_means.loc[:,use_pairs]
index = use_means.index
columns = use_means.columns
use_means = use_means.to_numpy()
use_means = pd.DataFrame((use_means - use_means.min(1).reshape((use_means.shape[0],1)))/(use_means.max(1).reshape((use_means.shape[0],1)) - use_means.min(1).reshape((use_means.shape[0],1))))
use_means.index = index
use_means.columns = columns

use_means['interaction'] = use_means.index
use_means = pd.melt(use_means,id_vars="interaction",value_name="value_mean")
use_means.index = [i[0]+"_"+i[1] for i in zip(use_means['interaction'],use_means['variable'])]

cpdb_results['pvalues'].index = cpdb_results['pvalues']['interacting_pair']
use_pvalue = cpdb_results['pvalues'].loc[select_pairs['interaction_name']]
use_pvalue = use_pvalue.loc[:,use_pairs]
use_pvalue[use_pvalue < 0.001] = 0.001
use_pvalue = -np.log10(use_pvalue)
use_pvalue['interaction'] = use_pvalue.index
use_pvalue = pd.melt(use_pvalue,id_vars="interaction",value_name="value_p")
use_pvalue.index = [i[0]+"_"+i[1] for i in zip(use_pvalue['interaction'],use_pvalue['variable'])]

data = use_pvalue
data['value_mean'] = use_means.loc[data.index, 'value_mean']
data['circle'] = False
data.loc[data['value_p']>1.6, 'circle'] = True
data.head(5)
```

```python
fig,ax = plt.subplots()
fig.set_figwidth(12)
fig.set_figheight(8)
ax.set_xlabel("")
data['alpha'] = (data['value_p']+3)/(data['value_p']+3).max()
data['dot_size'] = data['value_mean'] * 120
scatter = ax.scatter(x=data['variable'],y=data["interaction"],s=data['dot_size'],c=data['value_p'],cmap="Spectral",alpha=data['alpha'],linewidth=0,edgecolors="black")
tmp = data.loc[data['circle'],:]
c = plt.cm.Reds(tmp['value_mean']/tmp['value_mean'].max())
scatter_circle = ax.scatter(x=tmp['variable'],y=tmp["interaction"],s=tmp['dot_size'],facecolors="none",linewidth=1.,edgecolors=c)
ax.tick_params(axis='x', labelrotation=90)
ax.xaxis.set_ticks_position("top")
ax.xaxis.set_label_position("top")
fig.colorbar(scatter,ax=ax,label="-log10 pvalue")

ax2 = fig.add_axes([1.2,0.3,0.5,0.08])
diff = data['dot_size'].max() - data['dot_size'].min()
step = diff/4
size_range = np.arange(data['dot_size'].max(), data['dot_size'].min(), step * -1)[::-1].tolist()
size_range = [float(data['dot_size'].min())] + size_range 
ax2.scatter(x=np.arange(len(size_range))+0.5,y=np.zeros(len(size_range)),s=size_range,c='white',linewidth=2.,edgecolors="black",cmap="Reds")
ax2.grid(False)
ax2.set_xticks(np.arange(len(size_range)) + 0.5)
diff = data['value_mean'].max() - data['value_mean'].min()
step = diff/4
labels = np.arange(data['value_mean'].max(), data['value_mean'].min(), step * -1)[::-1].tolist()
labels = [float(data['value_mean'].min())] + labels
labels = [str(i)[:str(i).find(".")+2] for i in labels]
ax2.set_xticklabels(labels, fontsize="small")
ax2.set_ylim(-0.1,0.1)
ax2.tick_params(axis="y", left=False, labelleft=False, labelright=False)
ax2.set_title("scaled significant means")
ax2.set_xlim(0,5)
fig.savefig("5.3.FigureS3A.pdf",format="pdf",bbox_inches="tight")
fig.show()
```
